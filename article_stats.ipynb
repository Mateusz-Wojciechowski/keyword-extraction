```python
import pandas as pd
import pickle
```

```python
with open('article_lengths.pkl', 'rb') as file:
        articles = pickle.load(file)
```

```python
type(articles)
```

```python
data = []
for topic, lengths in articles.items():
    for length in lengths:
        data.append({"Topic": topic, "Length": length})
```

```python
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
df.boxplot(column='Length', by='Topic', grid=False)
plt.title('Article length distribution')
plt.suptitle('') 
plt.xlabel('Topic')
plt.ylabel('Article length')

plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
article_counts = df['Topic'].value_counts()

plt.figure(figsize=(12, 6))
article_counts.plot(kind='bar')
plt.title('Amount of articles in each topic')
plt.xlabel('Topic')
plt.ylabel('Amount of articles')
plt.show()
```
